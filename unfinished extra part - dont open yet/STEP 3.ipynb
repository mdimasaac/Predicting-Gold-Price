```python
import requests
import pandas as pd
import numpy as np
```

```python
def to_float(x):
    if "," in str(x):
        x = x.replace(",","")
    return round(float(x),2)
```

```python
data = pd.read_csv("data_raw.csv")
data = data.drop(data.columns[0], axis = 1)
display(data.shape)
display(data.head())
data = data.drop_duplicates()
```

```python
data_prices = data.drop(["date", "year"], axis = 1)
data_time = data[["date","year"]].copy()
data_prices = data_prices.fillna(value = np.nan)
# data_prices["palladium"] = data_prices["palladium"].astype(float)
# data_prices["platinum"] = data_prices["platinum"].astype(float)
remove_comma = ["dji","djt","snp","palladium","platinum"]
for i in remove_comma:
    data_prices[i] = data_prices[i].apply(to_float)
data_prices = data_prices.interpolate()
data_prices = data_prices.round(2)
data = pd.concat([data_time, data_prices], axis = 1)
data["date"] = data["date"].astype("datetime64[ns]")
```
